### Filtrar

```python
df_2 = df[df.asset_id == 2]
df_2
```

```python
df_2.asset_id.nunique()
```

```python
df_3 = df[df.asset_id.isin([2, 70, 205])]
df_3
```

```python
df_3.asset_id.nunique()
```

```python
# | = "o"
# & = "y"
# ~ = negación
df_4 = df[(df.asset_id.isin([2, 70, 205])) & (df.Temperature_C >= 24)]
df_4
```

```python
df_4.asset_id.nunique()
```

```python
df_4.Temperature_C.min()
```

```python
df_5 = df[(df.sitio == 'Store') | (df.Temperature_F >= 76)]
df_5
```

```python
df_5.sitio.nunique()
```

```python
df_6 = df[(df.Temperature_F >= 76) & (df.Temperature_F <= 80)]
df_6
```
