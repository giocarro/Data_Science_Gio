# **Lugar menos poblado y mas poblado**

```python
# Lugar menos poblado
df1.loc[df1['total.pop'] == min(df1['total.pop'])]
```

```python
# Lugar mas poblado
df1.loc[df1['total.pop'] == max(df1['total.pop'])]
```

```python
df_tp = df1[['name', 'total.pop']].sort_values('total.pop', ascending = False)
df_tp
```

```python
fig_tp = px.histogram(df_tp, x='name', y='total.pop')
fig_tp.update_layout(
    xaxis_title = 'Condados', # Título Eje X
    yaxis_title = 'Poblacion', # Titulo Eje Y
    title_text = 'Poblacion de USA por condados', # Título
    )
fig_tp.show()
```

# Lugar con el menor y mayor numero de personas de raza blanca

## Condados con mas del 50% de gente de raza blanca

```python
df_wt = df1[['name','total.pop','white']]
df_wt
```

```python
df_wt1 = df_wt.loc[df_wt['white'] > 50].sort_values('white', ascending = False)
df_wt1
```

```python
#Cantidad de condados con mas del 50% con gente de raza blanca
len(df_wt1['name'])
```

```python
#Cantidad de condados en total sin datos nulos
len(df1['name'])
```

```python
#Porcentaje de condados donde mas del 50% de la poblacion es de raza blanca
len(df_wt1['name'])/len(df1['name'])*100
```

### Gráfica de barras de los condados con mas del 50% de poblacion de raza blanca.

```python
fig_wt1 = px.bar(df_wt1, x='name', y='total.pop', color='white')
fig_wt1.update_layout(
    xaxis_title = 'Condados', # Título Eje X
    yaxis_title = 'Poblacion', # Titulo Eje Y
    title_text = 'Condados de USA donde mas del 50% de la poblacion es de raza blanca', # Título
    )
fig_wt1.show()
```
